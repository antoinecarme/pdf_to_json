```python
import pdf_to_json as p2j

from IPython.core.display import display, HTML
```

```python
url1 = "https://www.ohchr.org/EN/UDHR/Documents/UDHR_Translations/eng.pdf"

def process_uri(uri):
    lConverter = p2j.pdf_to_json.pdf_to_json_converter()
    lDict = lConverter.convert(uri)
    print(lDict["MetaData"])
    for p in lDict["Pages"].keys():
        print([lDict["Pages"][p][x] for x in ['Label', 'Index', 'Size', 'Attributes']])
        display(HTML(lDict["Pages"][p]["Image"]))
        lines = lDict["Pages"][p]["Text"].split("\n")
        print("\n".join(lines[:5]) + "\n...\n" + "\n".join(lines[-5:]) + "\n\n\n")
```

```python
process_uri(url1)
```
